## Break & Continue

### Break

Break, döngünün herhangi bir yerinde veya zamanda durmasını sağlar. 

```python
a=0
while a<20:
    print(a)
    a=a+1
```

```python
i=0
while i<20:
    print(i)
    if i==15:
        break
    i=i+1
```

"Break" sadece içinde bulunduğu döngüyü sona erdirir. 

```python
while True:
    print("Merhaba")
```

```python
while True:
    print("Merhaba")
    çıkış=input("Çıkmak için q harfine basınız.")
    if çıkış=="q":
        print("Programdan çıkış yapılıyor...")
        break
```

### Continue

Komutun herhangi bir yerinde, blok un başına dönmek için kullanılır.

```python
listem1=list(range(50))
```

```python
print(listem1)
```

```python
for i in listem1:
    print("İ'nin değeri:", i)
```

```python
for i in listem1:
    if i%2==0:
        continue
    print("İ'nin değeri:", i)
```

### Faktöriyel

0!=1 ve 1!=1
2!=1*2 
3!=1*2*3

```python
def faktoriyel(sayi):
    faktor=1
    if sayi==0 or sayi==1:
        print("Sayının faktöriyeli:", faktor)
    else:
        while sayi>=1:
            faktor=faktor*sayi
            sayi=sayi-1
        print("Sayının faktöriyeli:", faktor)
            
    
```

```python
faktoriyel(0)
```

```python
faktoriyel(1)
```

```python
faktoriyel(10)
```

```python
faktoriyel(100)*2
```

### Fonksiyonlarda esnek sayıda parametre kullanımı

Bunun için * sembolünü kullanılır.

```python
def çarpım(*sayilar):
    sonuc=1
    for sayi in sayilar:
        sonuc=sonuc*sayi
    return sonuc
```
